```python
from bs4 import BeautifulSoup as bfS
import requests
import pandas as pd
```

```python
url="https://www.worldometers.info/world-population/population-by-country/"
#fetching the url
webpage=requests.get(url)
webpage
```

```python
soup=bfS(webpage.text,'html') #parsing it into html format
soup
```

```python
popTable=soup.find('table') #getting the table
```

```python
#getting the table headers
tblHeaders=[header.text.strip() for header in popTable.find_all('th')]
tblHeaders
```

```python
#getting the row data and data values from each row
row_data=popTable.find_all('tr')[1:]
countryPop=[]
for row in row_data:
    values=row.find_all('td')
    eValue=[v.text.strip() for v in values]
    countryPop.append(eValue)


print(countryPop)
```

```python
df=pd.DataFrame(countryPop,columns=tblHeaders)
df
```

```python
#creating csv file
df.to_csv('WorlPopulation.csv',index=False)
print('File created successfully')
```
