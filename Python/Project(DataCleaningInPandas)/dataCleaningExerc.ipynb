```python
import pandas as pd
dirtyDf=pd.read_csv('work.csv')
dirtyDf
```

```python
#removing spaces and renaming
dirtyDf['Name']=dirtyDf['Name'].str.strip()
dirtyDf['Department']=dirtyDf['Department'].str.strip()
dirtyDf.columns=dirtyDf.columns.str.replace(' ','_')
dirtyDf['Joining_Date']=dirtyDf['Joining_Date'].str.replace('-','/')
print(dirtyDf.to_string())
```

```python
#consistent formatting
dirtyDf['Name']=dirtyDf['Name'].str.lower()
dirtyDf['Department']=dirtyDf['Department'].str.lower()#string formats
dirtyDf['Salary'][1]=60000 #, please teach me how to fix this one(when the is a string in a coloumn that is supposed to contain integers) withouth having to go through the intire dataset,just in case of big datasets
dirtyDf['Salary']=dirtyDf['Salary'].astype(float)  #number formats
dirtyDf['Joining_Date'][7]='NaT'
dirtyDf['Joining_Date'] = pd.to_datetime(dirtyDf['Joining_Date'], errors='coerce', dayfirst=True)


dirtyDf
```

```python
#removing duplicates
print(f'\n\n Duplicated rows:\n{dirtyDf[dirtyDf.duplicated()]}')

dirtyDf=dirtyDf.drop_duplicates()
print(f'\nDuplicates Dropped\n{dirtyDf.to_string()}')
```

```python
#dealing with null value
#name coloumn
print('row with an empty name:')
print(dirtyDf[dirtyDf['Name'].isnull()])
print()
dirtyDf.dropna(subset='Name',inplace=True)
print('Row deleted(not useful to the data,no primary key)')
print(f'\n{dirtyDf.to_string()}')
```

```python
#Age coloumn
dirtyDf['Age']=dirtyDf['Age'].str.replace(' ','NaN')
dirtyDf['Age']=dirtyDf['Age'].astype(float)
y=round(dirtyDf['Age'].mean(),2)
dirtyDf['Age'].fillna(y,inplace=True)
print(dirtyDf)
```

```python
#Department
dirtyDf['Department'].fillna('Unstated',inplace=True)
print(dirtyDf)
```

```python
dirtyDf['Age']=dirtyDf['Age'].astype(int)
print(dirtyDf)
```
