```python
training_df = pd.DataFrame({
    'EmployeeID': [1001, 1002, 1005, 1010, 1012],
    'TrainingCompleted': ['Yes', 'No', 'Yes', 'No', 'Yes']
})
```

```python
df=df.merge(training_df,on='EmployeeID',how='left')
```

```python
df[df['TrainingCompleted']=='Yes']
df = df.drop('TrainingCompleted', axis=1)
```

```python
dfFilt=df[(df['Department']=='Marketing')&(df['YearsExperience']>=5)]
dfSort=dfFilt.sort_values(by=['YearsExperience','Salary'],ascending=[False,False])
dfSort
```

```python
# Select top 10 from the sorted, filtered data
top10 = dfSort.head(10)

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='EmployeeID', y='Salary', data=top10, palette='viridis')
plt.title('Top 10 Marketing Employees (â‰¥5 Years Experience) by Salary')
plt.xlabel('Salary')
plt.ylabel('Employee ID')
plt.tight_layout()
plt.show()
```

```python
df.filter(items=['EmployeeID','Department','Salary','YearsExperiance'])
```

```python
df['HireDate']=pd.to_datetime(df['HireDate'])
df['HireDate'].info()
```

```python
df['MonthOfHire']=df['HireDate'].dt.strftime('%B')
df['YearOfHire']=df['HireDate'].dt.year
```

```python
df
```

```python
lDf=df.groupby(['YearOfHire','MonthOfHire'],as_index=False).agg({'EmployeeID':'count'})
months_of_year=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['MonthOfHire']=pd.Categorical(df['MonthOfHire'],categories=months_of_year,ordered=True)
lDf.sort_values(by=['YearOfHire','MonthOfHire'],inplace=True)
```

```python
lDf
```

```python
plt.style.use('tableau-colorblind10')
lDf[lDf['YearOfHire']==2019].plot(kind='line',y='EmployeeID',x='MonthOfHire',title='Employees Hired Over The Year Of 2019',ylabel='Number Of Employees',xlabel='Months(2019)',color='green')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=lDf, x='MonthOfHire', y='EmployeeID', hue='YearOfHire', marker='o')
plt.title('Monthly Hiring Trends by Year')
plt.xlabel('Month')
plt.subplots()
plt.ylabel('Employees Hired')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Select relevant numerical columns
num_cols = ['Age', 'YearsExperience', 'Salary']  # Adjust as needed
corr_matrix = df[num_cols].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Employee Metrics")
plt.tight_layout()
plt.show()
```
