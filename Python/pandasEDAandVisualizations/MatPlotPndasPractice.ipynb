```python
df['OrderDate']=pd.to_datetime(df['OrderDate']) #converting to date time 
df['months'] = df['OrderDate'].dt.strftime('%B') #getting the months only


months_ordered=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['months']=pd.Categorical(df['months'],categories=months_ordered,ordered=True)    #arranging the months in order

dfSales=df.groupby('months').agg({'TotalPrice':'sum'}) 
dfSales
dfSales.plot(kind='line',color='orange')
plt.title('Monthly Sales Trend(total sales per month)')
plt.ylabel('TotalSales')
plt.xlabel('Month')
plt.show()
```

```python
scDf=df.groupby('Quantity').agg({'TotalPrice':'sum'})
sns.scatterplot(x='Quantity',y='TotalPrice',data=scDf)#can i get the line/trend line between the dots
plt.show()#how do I calculate the correlation between the two variables
correlation = df['Quantity'].corr(df['TotalPrice'])
print(f'Correlation between Quantity and TotalPrice: {correlation:.2f}')
```

```python


# Step 1: Select relevant numeric columns
numeric_cols = df[['Quantity', 'UnitPrice', 'TotalPrice']]

# Step 2: Compute the correlation matrix
corr_matrix = numeric_cols.corr()

# Step 3: Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add title
plt.title('Correlation Heatmap of Sales Metrics')
plt.tight_layout()
plt.show( )
```
