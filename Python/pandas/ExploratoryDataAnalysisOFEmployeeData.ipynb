```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('employee_data (1).csv')
df.head(5)
```

```python
print(df.head(5))# first 5 rows
print()
print('Rows and Columns(Shape)')
print(df.shape)
print()
print('Table info')
print(df.info())
```

```python
# Descriptive statistics for numerical columns
print(df.describe())

# Value counts for the 'Attrition' column
print('\nValue counts in the Attrition column:')
print(df['Attrition'].value_counts())

# Unique values in each column
print('\nNumber of unique values in each column:')
print(df.nunique())
```

```python
df
```

```python
dfStats=df.groupby('Age',as_index=False).agg({'MonthlyIncome':'mean'}) # getting the average salary per age group
sns.scatterplot(data=dfStats,x='MonthlyIncome',y='Age',color='orange')#how do I get a line to show between the points
plt.title('Relationship between Age and MonthlyIncome')
plt.show()
```

```python
sns.regplot(data=df, x='Age', y='MonthlyIncome', scatter_kws={'alpha':0.4}, line_kws={'color':'red'})
plt.title('Regression Line: Age vs MonthlyIncome')
plt.show()
```

```python
sns.countplot(data=df,x='Attrition')
plt.title('Attrition(No vs Yes)')
plt.show()

sns.histplot(data=df,x=df['MonthlyIncome'],kde=True,color='green',bins=30)
plt.title('Distribution of MonthlyIncomes of Employees')
plt.show()

sns.boxplot(data=df,x='Age')
plt.title('Spread Of Different Age groups')
plt.show()
```
