```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
df = sns.load_dataset('tips')  # Restaurant bill tips dataset
df
```

```python
sns.barplot(x='day', y='total_bill', data=df)
plt.title('Average Bill per Day')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
#plt.title('Tip vs Total Bill')
#plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Bill Distribution by Day')
plt.show()
```

```python
df=pd.read_csv("Downloads/monthly_sales.csv")
df
```

```python
xAxis=df['Month']
yAxis=df['Sales']
plt.hist(yAxis,bins=5,color='green',edgecolor='black')
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales($)')
plt.grid(True)
plt.show()
```

```python
data = {
    'HR': [40000, 42000, 45000, 47000, 49000],
    'IT': [60000, 62000, 65000, 68000, 70000],
    'Sales': [50000, 51000, 53000, 55000, 56000]
}

plt.boxplot(data.values(), labels=data.keys())
plt.title('Salary Distribution by Department')
plt.ylabel('Salary ($)')
plt.grid(True)
plt.show()
```
