```python
import csv
import datetime
import pandas as pd
import os

file_path="MyBooks.csv"


class Book:
     MyBooks=[["Title","Author","ISBN No","Availability"]]
     num_of_Books=0
    # creating the file
     if os.path.exists(file_path):
         pass
     else:
         myFile=open(file_path,"w",newline="")
         addBook=csv.writer(myFile)
         addBook.writerow(MyBooks[0])
         myFile.close()

    
     def __init__(self,title,author,isbn,available):
         self.title=title
         self.author=author
         self.isbn=isbn
         self.available=available
         Book.MyBooks.append([self.title,self.author,self.isbn,self.available])  #adding each book to the list
         Book.num_of_Books+=1
         


     def __str__(self):
         return f"Name:{self.title}{'\n'}Author:{self.author} {'\n'}ISBN No:{self.isbn} {'\n'}"

     def addBook(self):

         #adding the book to the file
         myFile=open(file_path,"a",newline="")
         addBook=csv.writer(myFile)
         i=len(Book.MyBooks)-1
         addBook.writerow(Book.MyBooks[i])
         myFile.close()
         
patreon_file='MyPatreons.csv'
class Patreon:
    myPatreons=[['Patreon_ID','Name','Surname','Cellphone','Address']]
    patreonIDs=[]
    num_of_patreons=0
    #creating the patreon file
    if os.path.exists(patreon_file):
        pass
    else:
       myPatreonFile=open(patreon_file,'w',newline='')
       writeMe=csv.writer(myPatreonFile)
       writeMe.writerow(myPatreons[0])    
       myPatreonFile.close()
    

    
    def __init__(self,patreonId,name,surname,cellphone,address):
        self.patreonId=patreonId
        self.name=name
        self.surname=surname
        self.cellphone=cellphone
        self.address=address
        Patreon.myPatreons.append([self.patreonId,self.name,self.surname,self.cellphone,self.address])
        Patreon.num_of_patreons+=1
        Patreon.patreonIDs.append([self.patreonId,self.name,self.surname,])

    def __str__(self):
        return f"PatreonID:{self.patreonId}{'\n'}Name:{self.name.capitalize()} {self.surname.capitalize()}{'\n'}Cell Phone:{self.cellphone}{'\n'}Address:{self.address}"
 
    def addPatreon(self):
        #adding each patreon to the file
        myPatreonFile=open(patreon_file,'a',newline='')
        writeMe2=csv.writer(myPatreonFile)
        i=len(Patreon.myPatreons)-1
        writeMe2.writerow(Patreon.myPatreons[i])
        myPatreonFile.close()
        print(f"{self.name.capitalize()} {self.surname.capitalize()} has been added to the sytem")

    #searching if a patreon is in the system
    @classmethod
    def searchPatreon(cls,mypatreonID,patreonName):
       ptr=0
       i=len(cls.patreonIDs)-1
       while ptr<=i:
           if mypatreonID==cls.patreonIDs[ptr][0] and patreonName.lower()==cls.patreonIDs[ptr][1].lower():
               return True
           else:
               ptr+=1

       return False



myBorrowedBooks="BorrowedBook.csv"#creating a file for borrowed books
class Library:
    availableBooks=[]
    borrowedBooks=[['Book Name','ISBN No.','Date','Returned','PateronID']]
    if os.path.exists(myBorrowedBooks):
        pass
    else:
        myFile=open(myBorrowedBooks,"w",newline="")
        addBook=csv.writer(myFile)
        addBook.writerow(borrowedBooks[0])
        myFile.close()

    def __init__(self,bookname,patreonMamber):
        self.bookname=bookname
        self.patreonMamber=patreonMamber



    @classmethod
    def searchBook(cls,bookName):
        #fetching the books from the file
        books="MyBooks.csv"
        file=open(books,"r")
        content=csv.reader(file)
        for line in content:
            cls.availableBooks.append(line)      
        file.close()

        #searching if book is avilable
        bookControl=len(cls.availableBooks)-1
        i=0
        loopMe=True
        found=False
        while (i<=bookControl) and loopMe :
            if cls.availableBooks[i][0].lower()==bookName.lower():
                 found=True
                 loopMe=False
            elif  str(cls.availableBooks[i][3])=='False':
                print(f'Book Name:{cls.availableBooks[i][0]} is not in the shelfs, currently unavailable/borrowed')
                found=True
                pass
            else:
                 found=False 
            i+=1
        if found:
            print(f'{bookName} is in the system')
            print("List of books in the system:")
            print()
            df=pd.read_csv('MyBooks.csv')
            print(df.to_string())
            print()
            #for books in cls.availableBooks:
                 # print(books[0],'          ',books[2])
           
              
        else:
            print(f"{bookName} is NOT in the system!")
            print("List of books in the system:")
            print()
            df=pd.read_csv('MyBooks.csv')
            print(df.to_string())
            print()
                
    def borrowBook(self):
        Library.searchBook(self.bookname)
        print()
        i=len(Library.availableBooks)-1
        u=0
        found=False
        print("PLEASE ENTER the details of the patreon member whose here to borrow a book")
        is_id=True
        while is_id:
            patreonId=input('Please enter the patreonID(numbers only):')
            if patreonId.isdigit():
                is_id=False
                patreonId=int(patreonId)
            else:
                is_id=True
                print('PLEASE ENTER DIGITS ONLY FOR PATREON ID')
                
        patreonName=input('Please enter the patreon member(First name only):')
        is_Patreon=Patreon.searchPatreon(patreonId,patreonName)   #search for the the memember in patreon table in order to borrow the book
        if is_Patreon:                                            # allow the user to borrow the book if they are a patreon
          num_of_borrowed_books=len(Library.borrowedBooks)-1
          x=num_of_borrowed_books
          temp_list_books=[]
          while u<=i:
                if self.bookname.lower()==Library.availableBooks[u][0].lower() and Library.availableBooks[u][3]=='True' :
                    temp_list_books.append(Library.availableBooks[u][0])
                    temp_list_books.append(Library.availableBooks[u][2])
                    temp_list_books.append(datetime.datetime.now())
                    temp_list_books.append('False')
                    temp_list_books.append(patreonId)
                    Library.borrowedBooks.append(temp_list_books.copy())
                    temp_list_books=[]
                    Library.availableBooks[u][3]='False'
                    found=True
                else:
                   pass
                 
                u+=1
       
          books="MyBooks.csv"
          file=open(books,"w",newline="")
          content=csv.writer(file)
          for line in Library.availableBooks:
               content.writerow(line)      
          file.close()
          print()
          if found:
               print(f'{self.bookname} has been borrowed this has been updated in the sytem')
               myBorroweds=open(myBorrowedBooks,'a',newline='')
               writeMe2=csv.writer(myBorroweds)
               i=len(Library.borrowedBooks)-1
               writeMe2.writerow(Library.borrowedBooks[i])
               myBorroweds.close()
                
          else:
                print(f'{self.bookname} is currently not available in the sytem')
            
          Library.availableBooks=[]

        else:
            print(f'User{'\n'}Name:{patreonName}{'\n'}PatreonID:{patreonId}{'\n'} is not a patreon member therfore they can borrow a book' )
        
        
```
