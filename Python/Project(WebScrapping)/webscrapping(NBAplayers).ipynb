```python
from bs4 import BeautifulSoup as btfS
import requests 
import pandas as pd
```

```python
url="https://www.basketball-reference.com/leagues/NBA_2024_per_game.html"
webpage=requests.get(url)
webpage
```

```python
soup=btfS(webpage.text,"html.parser")
```

```python
playersTbl=soup.find('table')
```

```python
rows=playersTbl.find_all('tr')[1:]
dataList=[]
for row in rows:
    dataValues=row.find_all(['td','th'])
    rowList=[values.text.strip() for values in dataValues]
    dataList.append(rowList)

```

```python
tblHeaders=playersTbl.find_all('tr')[0]
headers=[top.text.strip() for top in tblHeaders.find_all('th')]
headers
```

```python
df=pd.DataFrame(dataList,columns=headers)
df
```

```python
df.to_csv('NBA_PlayerStats.csv',index=False)
print('File created')
```
