```python
from bs4 import BeautifulSoup as bSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_highest-grossing_films"  
```

```python
webpage=requests.get(url)
```

```python
soup=bSoup(webpage.text,'html')
```

```python
myTbl=soup.find_all('table')[0]
```

```python
#num=len(soup.find_all('th'))
lst=[x.text.strip() for x in soup.find_all('td')[:5]]
print(lst)
```

```python
titles=[x.text.strip() for x in myTbl.find('tr').find_all(['td','th'])]
print(titles)
```

```python
clData=myTbl.find_all('tr')
print(clData[1].find_all(['td','th'])[2].text.strip())
```

```python
data=[]
for row in clData:
    row_data=row.find_all(['td','th'])
    individual_rwData=[x.text.strip() for x in row_data]
    data.append(individual_rwData)
    print(individual_rwData)
```

```python
mydf=pd.DataFrame(data[1:], columns=data[0])
mydf
```

```python
mydf.to_csv('TopGrossingFilms.csv',index=False)
```
